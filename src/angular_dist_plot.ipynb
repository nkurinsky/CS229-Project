```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
```

```python
data1 = np.load("unprocessed_angular_dist.npy")
data2 = np.load("angular_dist.npy")
truth = np.load("truth.npy")
```

```python
plt.figure()
_ = plt.hist(data1[truth==1],bins=20,label="Galaxies",color='b')
_ = plt.hist(data1[truth==2],bins=20,label="Stars",color='r')
plt.legend()
```
